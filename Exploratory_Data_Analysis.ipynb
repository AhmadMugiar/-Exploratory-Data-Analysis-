#DATA CLEANING

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


data = pd.read_csv('/content/project_1.csv')
data
```

```python
drp = data.drop(columns=["remote_ratio","salary_currency","employee_residence","salary"])
drp
```

```python
drp['salary_in_usd'] = drp['salary_in_usd'].astype(int)
drp
```

```python
cntry = drp.loc[drp['company_location'].isin(['US', 'CA','ES'])]
cntry
```

```python
job = cntry.loc[cntry['job_title'].isin(["Data Engineer","Data Analyst","Data Scientist"])]
job
```

```python
employment = job.loc[job['employment_type'].isin(["FT","C"])]
employment
```

```python
experience = employment.loc[employment['experience_level'].isin(["SE"])]
experience
```

```python
company = experience.loc[experience['company_size'].isin(["M"])]
company
```

```python
rp = 16000
company["salary_in_usd"] =company["salary_in_usd"] * rp
company
```

```python
company
```

```python
dff = company.rename(columns={'salary_in_usd': 'salary_in_rupiah'})
dff
```

```python
jobb = dff.loc[dff['job_title'].isin(["Data Scientist","Data Analyst","Data Engineer"])]
jobb
```

```python
year = jobb.loc[jobb['work_year'].isin([2022,2023,2024])]
year.sort_values("salary_in_rupiah")
```

```python
list(year.job_title.unique())
list(year.work_year.unique())
list(jobb.experience_level.unique())
list(jobb.company_location.unique())
list(jobb.company_size.unique())
```

#PIVOT TABLE RATA RATA

```python
dp = pd.DataFrame(year)
dp
```

```python
pivot_table = pd.pivot_table(
    dp, values="salary_in_rupiah", index=["company_location","job_title","employment_type"], columns="work_year", aggfunc=["sum","mean","count"],fill_value=0)
pivot_table
```

```python
pivot_table.plot.barh(stacked = True)
```

```python
mean_2022 = pivot_table["mean", 2022]
mean_2023 = pivot_table["mean", 2023]
percentage_change = ((mean_2023 - mean_2022) / mean_2022) * 100
percentage_change_rounded_up = np.ceil(percentage_change)
pivot_table['PERSENTASE 2022 TO 2023'] = percentage_change_rounded_up
pivot_table['PERSENTASE 2022 TO 2023'] = pivot_table['PERSENTASE 2022 TO 2023'].apply(lambda x: f"{float(x)}%")
pivot_table
```

```python
mean_2023 = pivot_table["mean", 2023]
mean_2024 = pivot_table["mean", 2024]
percentage_change = ((mean_2024 - mean_2023) / mean_2023) * 100
percentage_change_rounded_up = np.ceil(percentage_change)
pivot_table['PERSENTASE 2023 TO 2024'] = percentage_change_rounded_up
pivot_table['PERSENTASE 2023 TO 2024'] = pivot_table['PERSENTASE 2023 TO 2024'].apply(lambda x: f"{float(x)}%")

pivot_table
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mengonversi pivot table ke format yang lebih cocok untuk plotting
pivot_table_reset = pivot_table.reset_index()

# Plotting data
plt.figure(figsize=(14, 8))
sns.barplot(x='job_title', y='PERSENTASE 2022 TO 2023', hue='company_location', data=pivot_table_reset)

plt.xlabel('Job Title')
plt.ylabel('Persentase Perubahan 2022 ke 2023')
plt.title('Persentase Perubahan Mean Gaji dari 2022 ke 2023 per Job Title dan Negara')
plt.xticks(rotation=360)
plt.legend(title='Negara')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=pivot_table, x= "job_title", y= ("PERSENTASE 2022 TO 2023"), hue='job_title')
plt.title('Perbandingan Pendapatan Rata-Rata Berdasarkan Negara dan Profesi Tahun 2024')
plt.xlabel('job_title')
plt.ylabel('PERSENTASE 2022 TO 2023')
plt.show()
```
